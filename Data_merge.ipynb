```python
import seaborn as sns  
import matplotlib.pyplot as plt  
  
# 你有一个DataFrame df  
df = pd.read_csv('/home/zengzw/tmp/all_merge.csv')  # 从CSV文件读取数据  
plt.figure(figsize = (10,6))
# 根据列'C'的值绘制条形图（假设列'C'是分类数据）  
sns.countplot(x='cancer type', hue = "data source", data=df, width=0.4)  
plt.title('Count plot of Cancer Type') 
plt.xlabel('Cancer Type', fontsize=14)
plt.xticks(rotation=90)  # x轴标签旋转90度
plt.ylabel('Count', fontsize=14)
plt.savefig("/home/zengzw/tmp/figures/barplot_by_cancer_type.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
df = pd.read_csv('/home/zengzw/tmp/all_merge.csv')

# 计算每个'cancer type'对于'C'的计数
count_df = df.groupby(['cancer type', 'data source']).size().unstack(fill_value=0)

# 绘制堆叠条形图
count_df.plot(kind='bar', stacked=True, figsize=(6,6))

plt.title('Stacked Bar Plot of Cancer Type by Data Source')
plt.xlabel('Cancer Type', fontsize=14)
plt.xticks(rotation=90)
plt.ylabel('Count', fontsize=14)

plt.savefig("/home/zengzw/tmp/figures/stacked_barplot_by_cancer_type_and_data_source.png")
plt.show()
```
